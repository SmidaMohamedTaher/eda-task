```python
def FNMV(df,column_name):#filling numerical messing values
    df[column_name] = df[column_name].fillna(df[column_name].mean())
```

```python
FNMV(df,"AttendanceRate")
FNMV(df,"StudyHoursPerWeek")
FNMV(df,"PreviousGrade")
FNMV(df,"ExtracurricularActivities")
FNMV(df,"FinalGrade")
```

```python
def FCMV(df,column_name):#filling category messing values
    df[column_name] = df[column_name].fillna(df[column_name].mode()[0])
```
